```python
from pystac_client import Client
from odc.stac import load
import xarray as xr
import numpy as np
import odc.geo # noqa
```

```python
catalog = "https://stac.staging.digitalearthpacific.org"
collection = "dep_s2_mangroves"
```

```python
# Coordinates grabbed from Google Maps

# Raviravi, VL, Fiji
ll = (-11.728,140.421)
ur = (-1.601,154.001)


bbox = (ll[1], ll[0], ur[1], ur[0])

# Find STAC items
client = Client.open(catalog)
items = client.search(collections=[collection], bbox=bbox).items()
items = [i for i in items]

print(f"Found {len(items)} items")
```

```python
config = {
    collection: {
        "assets": {
            "mangroves": {"data_type": "int16"}
        }
    }
}

data = load(
    items,
    bbox=bbox,
    bands=["mangroves"],
    chunks={"x": 500, "y": 500},  # Corrected syntax for chunks
    stac_cfg=config
)
data
```
